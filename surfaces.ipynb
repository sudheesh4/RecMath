```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
import numpy as np
import sympy as smp
```

```python

def getmesh(n,Num):
    xlim=[-1,n]
    ylim=[-n,n]
    px=np.linspace(xlim[0],xlim[1],Num)
    py=np.linspace(ylim[0],ylim[1],Num)
    xx,yy=np.meshgrid(px,py)
    return xx,yy


def sphere():
    global X,Y,Z
    X=(r*smp.cos(theta)*smp.sin(phi))
    Y=r*smp.sin(theta)*smp.sin(phi)
    Z=r*smp.cos(phi)
def torus():
    global X,Y,Z
    X=(r+2*smp.cos(phi))*smp.sin(theta)
    Y=(r+2*smp.cos(phi))*smp.cos(theta)
    Z=2*smp.sin(phi)
def ntorus(n,x,y,r,ax):
    fx=1
    for i in range(1,n):
        fx=fx*(x-i)*(x-i+1)
    gxy=(fx+ y**2)#z find
    hxyz=np.sqrt(-(gxy**2)+ r)
    hxyzM=-np.sqrt(-(gxy**2)+ r)
    rz=[hxyz,hxyzM]
    for i in range(len(rz)):
        tes=(~np.isnan(rz[i])).astype('float')
        ax.scatter(x[tes!=0],y[tes!=0],rz[i][tes!=0],alpha=0.2)
def elliptic(n,x,y,r,A,B,ax):
    gxy=(x**3 +A*x +B - y**2)#z find
    hxyz=np.sqrt(-(gxy**2)+ r)**1
    hxyzM=-np.sqrt(-(gxy**2)+ r)**1
    rz=[hxyz,hxyzM]
    for i in range(len(rz)):
        tes=(~np.isnan(rz[i])).astype('float')
        ax.scatter(x[tes!=0],y[tes!=0],rz[i][tes!=0],alpha=0.2)
def somefunction(n,x,y,r,A,B,ax):
    gxy=(np.sin(x+y)-np.cos(x*y)+1)#z find
    hxyz=np.sqrt(-(gxy**2)+ r)**2
    hxyzM=-np.sqrt(-(gxy**2)+ r)**2
    rz=[hxyz,hxyzM]
    for i in range(len(rz)):
        tes=(~np.isnan(rz[i])).astype('float')
        ax.scatter(x[tes!=0],y[tes!=0],rz[i][tes!=0],alpha=0.2)

def wineglass(x,z,r,ax):
    gxz=x**2 -b  -(np.log(z+3.2))**2
    h1=np.sqrt(-gxz)
    h2=-np.sqrt(-gxz)
    rz=[h1,h2]
    for i in range(len(rz)):
        tes=(~np.isnan(rz[i])).astype('float')
        ax.scatter(x[tes!=0],rz[i][tes!=0],z[tes!=0],alpha=0.2)
def elliptic2(n,x,y,r,A,B,ax):
    gxy=(x**3 +A*x +B - y**4)#z find
    hxyz=np.sqrt(-(gxy**2)+ r)
    hxyzM=-np.sqrt(-(gxy**2)+ r)
    rz=[hxyz,hxyzM]
    for i in range(len(rz)):
        tes=(~np.isnan(rz[i])).astype('float')
        ax.scatter(x[tes!=0],y[tes!=0],rz[i][tes!=0],alpha=0.2)

def ellipse(x,y,r,ax):
    gxy=(x**2 +y**2 -1+x*y)#z find
    hxyz=np.sqrt(-(gxy**2)+ r)
    hxyzM=-np.sqrt(-(gxy**2)+ r)
    rz=[hxyz,hxyzM]
    for i in range(len(rz)):
        tes=(~np.isnan(rz[i])).astype('float')
        ax.scatter(x[tes!=0],y[tes!=0],rz[i][tes!=0],alpha=0.2)
def ampersand(x,y,r,ax):
    gxy=(y**2 -x**2)*(x-1)*(2*x-3)-4*((x**2+y**2-2*x)**2)
    hxyz=np.sqrt(-(gxy**2)+ r)
    hxyzM=-np.sqrt(-(gxy**2)+ r)
    rz=[hxyz,hxyzM]
    for i in range(len(rz)):
        tes=(~np.isnan(rz[i])).astype('float')
        ax.scatter(x[tes!=0],y[tes!=0],rz[i][tes!=0],alpha=0.2)   

def devilcurve(x,y,r,a,b,ax):
    gxy=(y**2 * (y**2 -b**2))- (x**2 * (x**2- a**2))
    hxyz=np.sqrt(-(gxy**2)+ r)
    hxyzM=-np.sqrt(-(gxy**2)+ r)
    rz=[hxyz,hxyzM]
    for i in range(len(rz)):
        tes=(~np.isnan(rz[i])).astype('float')
        ax.scatter(x[tes!=0],y[tes!=0],rz[i][tes!=0],alpha=0.2)   
        
def hyperellip(x,y,r,ax):
    hx=0
    fx=x *(x+1)*(x+3)*(x-2)*(x+2)
    gxy=(y**2 -fx)
    hxyz=np.sqrt(-(gxy**2)+ r)
    hxyzM=-np.sqrt(-(gxy**2)+ r)
    rz=[hxyz,hxyzM]
    for i in range(len(rz)):
        tes=(~np.isnan(rz[i])).astype('float')
        ax.scatter(x[tes!=0],y[tes!=0],rz[i][tes!=0],alpha=0.2)     
```

```python
def torustest():
    ##Torus test
    Num=100
    b=10**-2

    fin=6
    start=4
    fig, ax = plt.subplots(1,fin-start,figsize=(10,10),subplot_kw=dict(projection='3d'))
    for n in range(start,fin):
        #n=5
        xx,yy=getmesh(n,n*Num)
        ntorus(n,xx,yy,b,ax[n-start])

def elliptictest():
    ##Elliptic test
    Num=1000

    xx=np.linspace(-10,10,Num)
    yy=np.linspace(-10,10,Num)
    xx,yy=np.meshgrid(xx,yy)
    fig, ax = plt.subplots(1,1,figsize=(7,7),subplot_kw=dict(projection='3d'))

    r=0.01

    elliptic(n,xx,yy,r,-1,0,ax)
    
def wineglasstest():
    ##Wineglass test
    Num=300
    fig, ax = plt.subplots(1,1,figsize=(7,7),subplot_kw=dict(projection='3d'))
    xx=np.linspace(-5,5,Num)
    zz=np.linspace(-5,20,Num)
    xx,zz=np.meshgrid(xx,zz)

    r=0.001
    wineglass(xx,zz,r,ax)
def elliptictest2():
    ##Elliptic test
    Num=10000

    xx=np.linspace(-10,10,Num)
    yy=np.linspace(-10,10,Num)
    xx,yy=np.meshgrid(xx,yy)
    fig, ax = plt.subplots(1,1,figsize=(7,7),subplot_kw=dict(projection='3d'))

    r=0.01

    elliptic2(n,xx,yy,r,-1,-1,ax)
def ellipsetest():
    ##Elliptic test
    Num=1000

    xx=np.linspace(-1,1,Num)
    yy=np.linspace(-1,1,Num)
    xx,yy=np.meshgrid(xx,yy)
    fig, ax = plt.subplots(1,1,figsize=(7,7),subplot_kw=dict(projection='3d'))

    r=0

    ellipse(xx,yy,r,ax)
    
def ampersandtest():
    ##Elliptic test
    Num=500

    xx=np.linspace(-0.5,2,Num)
    yy=np.linspace(-2,2,Num)
    xx,yy=np.meshgrid(xx,yy)
    fig, ax = plt.subplots(1,1,figsize=(7,7),subplot_kw=dict(projection='3d'))

    r=0.01

    ampersand(xx,yy,r,ax)


def deviltest():
    ##Elliptic test
    Num=100

    xx=np.linspace(-1,1,Num)
    yy=np.linspace(-1,1,Num)
    xx,yy=np.meshgrid(xx,yy)
    fig, ax = plt.subplots(1,1,figsize=(7,7),subplot_kw=dict(projection='3d'))

    r=0.01
    a=0.8
    b=1

    devilcurve(xx,yy,r,a,b,ax)

def hypertest():
    ##Elliptic test
    Num=1300

    xx=np.linspace(-4,4,Num)
    yy=np.linspace(-5,5,Num)
    xx,yy=np.meshgrid(xx,yy)
    fig, ax = plt.subplots(1,1,figsize=(7,7),subplot_kw=dict(projection='3d'))

    r=0.01
    hyperellip(xx,yy,r,ax)
    
```
