```python
import sympy as sp
import numpy as np
from IPython.display import display

```

```python
psi,xi,tau,epsilon,v,g=sp.symbols("psi xi tau epsilon v g")
```

```python
def fields(ch,i,j):
    if ch==0:
        return sp.symbols("theta_"+str(i)+str(j))
    elif ch==1:
        return sp.symbols("eta_"+str(i)+str(j))
    else:
        return sp.symbols("mu_"+str(i)+str(j))
def phase(n):
    return sp.symbols("lambda_"+str(n))

def Dps(f):
    global psi
    return sp.Derivative(f,psi)

def P(f,n):
    return (sp.Derivative(f,xi)+(1j*n*sp.symbols('k')*f/epsilon))
def P2(f,n):
    global xi,epsilon
    return (sp.Derivative(sp.Derivative(f,xi),xi)+(2*1j*n*sp.symbols('k')*sp.Derivative(f,xi)/epsilon) 
            -(n**2 * sp.symbols('k')**2 *f/(epsilon**2)))
def T(f,n):
    return (sp.Derivative(f,tau)+(1j*n*sp.symbols('Omega')*f/(epsilon**2)))

def getcoeff(expr,eps,harm):
    temp=sp.collect(sp.expand(expr),epsilon)
    return sp.collect(temp.coeff(epsilon,eps),sp.exp(1j*phase(harm))).coeff(sp.exp(1j*phase(harm)),1)
```

```python
bulk=0*psi
temp=0*psi
for n in range(-1,1+1):
    temp=temp+ (epsilon* Dps(fields(2,1,n)) * sp.exp(1j*phase(n)))
bulk=bulk+temp
#print(">>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>")
#display(temp)
temp=0*psi

for n in range(-2,2+1):
    temp +=  (epsilon**2 * Dps(fields(2,2,n)) * sp.exp(1j*phase(n)))
bulk=bulk+temp
#print(">>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>")
#display(temp)    
temp=0*psi

for n in range(-3,3+1):
    temp += (epsilon**3 * (Dps(fields(2,3,n))) * sp.exp(1j*phase(n)))
bulk=bulk+temp
#print(">>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>")
#display(temp)    
temp=0*psi

for n in range(-1,1+1):
    temp+= (epsilon**3 * P2(fields(0,1,n),n) * sp.exp(1j*phase(n)))
bulk=bulk+temp
#print(">>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>")
#display(temp)    
temp=0*psi

for n in range(-2,2+1):
    temp+= (epsilon**4 * P2(fields(0,2,n),n) * sp.exp(1j*phase(n)))
bulk=bulk+temp
#print(">>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>")
#display(temp)    
temp=0*psi

for n in range(-3,3+1):
    temp+= (epsilon**5 * P2(fields(0,3,n),n) * sp.exp(1j*phase(n)))
bulk=bulk+temp
#print(">>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>")
#display(temp)    
temp=0*psi

for n in range(-1,1+1):
    for m in range(-1,1+1):
        temp+= -2*(epsilon**4 * P(fields(1,1,n),n) * P(fields(2,1,m),m) * sp.exp(1j*phase(n+m)))
bulk=bulk+temp
#print(">>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>")
#display(temp)    
temp=0*psi

for n in range(-1,1+1):
    for m in range(-1,1+1):
       temp+=  - (epsilon**4 * P2(fields(1,1,n),n) * (fields(2,1,m)) * sp.exp(1j*phase(n+m)))
bulk=bulk+temp
#print(">>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>")
#display(temp)    
temp=0*psi

for n in range(-1,1+1):
    for m in range(-2,2+1):
        temp+= -2*(epsilon**5 * P(fields(1,1,n),n) * P(fields(2,2,m),m) * sp.exp(1j*phase(n+m)))
bulk=bulk+temp
#print(">>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>")
#display(temp)    
temp=0*psi

for n in range(-2,2+1):
    for m in range(-1,1+1):
        temp+= -2*(epsilon**5 * P(fields(1,2,n),n) * P(fields(2,1,m),m) * sp.exp(1j*phase(n+m)))
bulk=bulk+temp
#print(">>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>")
#display(temp)    
temp=0*psi

for n in range(-1,1+1):
    for m in range(-2,2+1):
        temp+= -(epsilon**5 * P2(fields(1,1,n),n) * (fields(2,2,m)) * sp.exp(1j*phase(n+m)))
bulk=bulk+temp
#print(">>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>")
#display(temp)    
temp=0*psi

for n in range(-2,2+1):
    for m in range(-1,1+1):
        temp+= -(epsilon**5 * P2(fields(1,2,n),n) * ((fields(2,1,m))) * sp.exp(1j*phase(n+m)))
bulk=bulk+temp
#print(">>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>")
#display(temp)    
temp=0*psi

for n in range(-1,1+1):
    for m in range(-1,1+1):
        for l in range(-1,1+1):
            temp+= +(epsilon**5 * P(fields(1,1,n),n) * P(fields(1,1,m),m) * (Dps(fields(2,1,l)))  * sp.exp(1j*phase(n+m+l)))
bulk=bulk+temp
#print(">>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>")
#display(temp)    
temp=0*psi
```
